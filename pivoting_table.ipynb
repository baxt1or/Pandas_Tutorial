```python
import pandas as pd 
import numpy as np
```

```python
# Sample Dummy Data to work with
df = pd.DataFrame({
    "Date": ["2024-01", "2024-01", "2024-01", "2024-02", "2024-02", "2024-02", 
             "2024-03", "2024-03", "2024-03"],
    "Region": ["North", "South", "East", "North", "South", "East", 
               "North", "South", "East"],
    "Product": ["A", "A", "A", "B", "B", "B", "C", "C", "C"],
    "Sales": [100, 150, 120, 130, 170, 140, 160, 180, 135]
})

df
```

```python
# Creating the function to Pivot the table
def pivot_table(df:pd.DataFrame, index, columns, values, aggfunc="mean"):
    
    items = df[columns].unique().tolist()

    agg_func = {
       item:(values, lambda x, item=item: x[df.loc[x.index, columns] == item].sum() if aggfunc == 'sum' else x[df.loc[x.index, columns] == item].count() if aggfunc=='count' else x[df.loc[x.index, columns] == item].mean())
       for item in items
    }

    result_df = df.groupby(index).agg(**agg_func)

    return result_df
```

```python
# Example Usage
pivot_table(df, index="Date", columns="Product", values="Sales", aggfunc="sum")
```

```python
df.pivot_table(index="Date", columns="Product", values="Sales", aggfunc="sum")
```
