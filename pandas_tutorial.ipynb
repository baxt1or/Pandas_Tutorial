# Merging and Concat

```python
data = [['LCHouse1', 1, 1], ['LCHouse1', 2, 10], ['LCHouse1', 3, 5], ['LCHouse2', 1, 2], ['LCHouse2', 2, 2], ['LCHouse3', 4, 1]]
warehouse = pd.DataFrame(data, columns=['name', 'product_id', 'units']).astype({'name':'object', 'product_id':'Int64', 'units':'Int64'})

data = [[1, 'LC-TV', 5, 50, 40], [2, 'LC-KeyChain', 5, 5, 5], [3, 'LC-Phone', 2, 10, 10], [4, 'LC-T-Shirt', 4, 10, 20]]
products = pd.DataFrame(data, columns=['product_id', 'product_name', 'Width', 'Length', 'Height']).astype({'product_id':'Int64', 'product_name':'object', 'Width':'Int64', 'Length':'Int64', 'Height':'Int64'})


result = pd.merge(warehouse, products, on="product_id", how="inner")

result
```

```python
data1 = {
    'A': [1, 2, 3],
    'B': [4, 5, 6]
}
df1 = pd.DataFrame(data1)


data2 = {
    'A': [7, 8, 9],
    'B': [10, 11, 12]
}
df2 = pd.DataFrame(data2)

result = pd.concat([df1, df2], axis=0)

result
```

```python
coffee_df = pd.read_csv("./coffee.csv")
```

```python
coffee_df.head()
```

```python
result = coffee_df.groupby("Day")["Units Sold"].sum().reset_index().sort_values(by="Units Sold", ascending=False)

result["rating"] = result["Units Sold"].rank(ascending=False, method="dense")

result
```

```python
result["sum_total"] = result["Units Sold"].cumsum()

result
```

```python
result["yesterday_sold"] = result["Units Sold"].shift(1)

result
```

```python
data = [[1, 'Joe', 85000, 1], [2, 'Henry', 80000, 2], [3, 'Sam', 60000, 2], [4, 'Max', 90000, 1], [5, 'Janet', 69000, 1], [6, 'Randy', 85000, 1], [7, 'Will', 70000, 1]]
employee = pd.DataFrame(data, columns=['id', 'name', 'salary', 'departmentId']).astype({'id':'Int64', 'name':'object', 'salary':'Int64', 'departmentId':'Int64'})
data = [[1, 'IT'], [2, 'Sales']]
department = pd.DataFrame(data, columns=['id', 'name']).astype({'id':'Int64', 'name':'object'})


df = pd.merge(employee, department, left_on="departmentId", right_on="id", how="inner")

df["rank"] = df.groupby("name_y")["salary"].transform("rank", method="min")

df[df["rank"] <= 3]
```
